```python
def calculate_ema(ema, data, smooth):
    for i in range(1,len(data)):
        ema = data[i]*(smooth/21)+ema*(1-(smooth/21))
return ema
```

```python
closing_ema = []
count = 0

for i in historical_data:
    close = []
    count += 1
    for j in i:
        close.append(j['close'])
    try:
        closing_ema.append(calculate_ema(close[0], close, 2))
    except IndexError:
        closing_ema.append(0)


    
#     print(close)
    
print(len(closing_ema))
# print(len(historical_data))
```
